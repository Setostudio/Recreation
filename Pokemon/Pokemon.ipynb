# Attribute analysis

```python
type_list = dataset['Type_1'].unique()
type_count = dataset[['Type_1']].groupby('Type_1')['Type_1'].count()

plt.figure(figsize=(16, 6))
ax = sns.barplot(sorted(type_count),type_list,palette="Reds_d")
ax.set_title('Number of type')
```

```python
type_index = dataset[['Type_1','HP','Attack','Defense','Sp_Atk','Sp_Def']].groupby(['Type_1'])['HP','Attack','Defense','Sp_Atk','Sp_Def'].mean()
type_index.head()
```

```python
type_index.idxmax()
```

```python
from math import ceil

def roundup(x):
    return int(ceil(x / 10.0)) * 10

def get_criterias(maximum_attribute):
    criteria_attribute = []
    for i in range(3):
        criteria_attribute.append(roundup(maximum_attribute * i /3))
    return criteria_attribute

maximum_attribute = roundup(type_index.max().max())
criteria_attribute = get_criterias(maximum_attribute)
criteria_attribute
```

```python
from math import pi

def plot_radar_comparision(type_name_1, type_name_2):
    attributes = type_index.columns
    attributes_count = len(attributes)

    type_1 = type_index.loc[type_name_1].values.tolist()
    type_1 += type_1[:1]

    type_2 = type_index.loc[type_name_2].tolist()
    type_2 += type_2[:1]
    
    angles = [n / float(attributes_count) * 2 * pi for n in range(attributes_count)]
    angles += angles [:1]

    angles2 = [n / float(attributes_count) * 2 * pi for n in range(attributes_count)]
    angles2 += angles2 [:1]

    ax = plt.subplot(111, polar=True)

    #Add the attribute labels to our axes
    plt.xticks(angles[:-1],attributes)
    plt.yticks([criteria_attribute[0],criteria_attribute[1],criteria_attribute[2]], [str(criteria_attribute[0]),str(criteria_attribute[1]),str(criteria_attribute[2])], color="grey", size=7)
    plt.ylim(0,120)
    #Plot the line around the outside of the filled area, using the angles and values calculated before
    ax.plot(angles,type_1,'o-', linewidth=2)
    ax.fill(angles, type_1, alpha=0.25)

    ax.plot(angles2,type_2)
    ax.fill(angles2, type_2, 'red', alpha=0.15)
    
    #Use figure text:
    plt.figtext(0.2,0.9,type_name_1,color="blue")
    plt.figtext(0.3,0.9,'Versus')
    plt.figtext(0.4,0.9,type_name_2,color="red")
    plt.show()
    plt.show()
```

```python
for key, value in type_index.idxmax().items():
    if (value != 'Dragon'):
        plot_radar_comparision(value, "Dragon")
```

```python
from sklearn.cluster import KMeans
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
dataset.head()
```

```python
from sklearn.model_selection import KFold
logistic_dataset = dataset[['isLegendary','HP','Attack','Defense','Sp_Atk','Sp_Def','Speed','Height_m','Weight_kg']]

X = logistic_dataset.iloc[:,1:9]
y = logistic_dataset['isLegendary']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state = 42)
```

```python
model = LogisticRegression(penalty = 'l2', fit_intercept = True, solver = 'lbfgs', max_iter = 1000, multi_class = 'ovr')
model
```

```python
model.fit(X_train, y_train)
```

```python
pred_labels = model.predict(X_train)
```

```python
pred_probs = model.predict_proba(X_train)
pred_probs
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

f1_score(y_train, pred_labels, average='weighted')  
```

```python
from sklearn.metrics import roc_curve

y_pred_prob = pred_probs[:, 1]

fpr, tpr, thresholds = roc_curve(y_train, y_pred_prob)
plt.plot(fpr, tpr, label = 'ROC')
plt.plot([0, 1], [0, 1], '--')
plt.legend(loc = 'upper center', fontsize = 'x-large')
plt.xlabel('FPR')
plt.ylabel('TPR')
```
